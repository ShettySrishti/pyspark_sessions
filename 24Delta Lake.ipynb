### PySpark API 

```python
# READ a particular version - PySpark API
dfsalesdelta = spark.read.table("salesdelta")  #spark.read.format("delta").load("path","/data/output/sales_hive_delta/")
display(dfsalesdelta)
```

- We can also read tables for its different versions as show below

```python

dfsalesdelta = spark.read.table("salesdelta@v0")  #spark.read.format("delta").load("path","/data/output/sales_hive_delta/")
display(dfsalesdelta)
```

```python
dfsalesdelta = spark.read.table("salesdelta@v0")  # This is version 0(v0). Version before UPDATE of 'price'
display(dfsalesdelta.where("product_name='Product D'"))
```

```python
dfsalesdelta = spark.read.table("salesdelta@v1")  # This is version 1(v1). Version after UPDATE of 'price'
display(dfsalesdelta.where("product_name='Product D'"))
```

```python
%sql
-- Read version in SQL queries
SELECT * FROm salesdelta@v1 WHERE product_name='Product D'
```

```python
# Schema evolution with mergeschema
dfnew = spark.sql("SELECT *, current_timestamp() AS time_now FROM salesdelta@v1 WHERE product_name='Product D'") # Adding new column time_now
display(dfnew)
```

```python
# APPEND data to existing delta table  -- appending a new record with a new column
dfnew.write.format("delta").mode("append").option("mergeSchema", True).option("path", "/data/output/sales_hive_delta/").saveAsTable("salesdelta")
```

```python
%sql
DESCRIBE HISTORY salesdelta  --New operation added as seen below for operation below
```

- As seen below, the '*mergeschema*' has added that column for all the records. 
- It has managed the Schema evolution by deafult which is also a benefit of Delta tables
- In case of parquet and hive tables, you would have to rewrite the whole data adding that column which is a disadvantage. 

```python
%sql
SELECT * FROM default.salesdelta -- WHERE product_name='Product D'
```

- What if we want to revert the version of the tables back to its old version? Can we do that? YES! 
- Let's see that below

```python
%sql
-- Reverting the table to its older version back

RESTORE TABLE salesdelta TO VERSION AS OF 1
```

```python
%sql
DESCRIBE HISTORY salesdelta -- You will see one mor transaction 'RESTORE' after the above operation
```

```python
%sql
SELECT * FROM salesdelta  --The restored version 1 (you can still see version 2 with @v2)
```

```python
display(dbutils.fs.ls("/data/output/sales_hive_delta"))
```

- In the above output, you can see many parquet files being wrriten
- This is because, everytime when you manipulate a version, Delta table tries to rewrite a different version of data in different Parquet files in order to maintain history

- Consider you have more then 100 versions of a particular table implying a lot of data right. 
- It is not wise to keep it as it is and we need to clean it up.
- In order to do that, Delta tables provide us with an option called **'*vacuum*'**
> Let's see that below

```python
dt = DeltaTable.forName(spark, "salesdelta")
dt.vacuum(0)
```

- In the above output, you can see '*Retention Period*' that has been provided
- Default retention log can be removed using '**spark.conf.set("spark.databricks.delta.retentionDurationCheck.enabled", False)**'


```python
spark.conf.set("spark.databricks.delta.retentionDurationCheck.enabled", False)
dt = DeltaTable.forName(spark, "salesdelta")
dt.vacuum(0)
```

IMPORTANT:
- vacuum(0): Removes all the older verions of the table and RETAINS ONLY the LATEST VERSION

```python
display(dbutils.fs.ls("/data/output/sales_hive_delta"))  # Now you will see only 9 parquet file all other unnecessary versions are removed
```

```python
%sql
-- Check what happened to the Metadata, versions are still available
DESCRIBE HISTORY salesdelta
```

All versions are still available. Let's check the table itself

```python
%sql
SELECT * FROM salesdelta  -- This is restored version so you will be able to see
```

- The above is the restored(latest) version, so you will be able to see only the latest versions.
- But now you wont be able to see other versions, it will throw an error.
- Let's confirm that below

```python
%sql
SELECT * FROM salesdelta@v2
```

> See, this failed because apart from restored version, all the verions of the tables are removed by 'vacuum'
