Let's convert it into Delta table to perform Z-Ordering.

```python
# Write the data in the form of Delta table

df = spark.read.csv("/databricks-datasets/definitive-guide/data/retail-data/all/online-retail-dataset.csv", inferSchema=True, header=True)
df.repartition(16).write.format("delta").mode("overwrite").option("path","/data/output/sales_delta_zordering/").saveAsTable("salesdelta")
```

```python
%sql
SELECT * FROM salesdelta
```

```python
# Data at delta location
display(dbutils.fs.ls("/data/output/sales_delta_zordering/"))  # You will see 16 files(excluding delta log) because we defined partition number as 16
```

```python
%sql
SELECT * FROM salesdelta WHERE InvoiceNo='540646'
```

```python
%sql
SELECT MIN(InvoiceNo), MAX(InvoiceNo), _metadata.file_name 
FROM salesdelta 
GROUP BY _metadata.file_name
ORDER BY MIN(InvoiceNo)

-- Before Z-Ordering Optimization
```

Z-Ordering

```python
spark.conf.set("sparks.databricks.delta.optimize.maxFileSize", 64*1024*8)
```

```python
%sql
OPTIMIZE salesdelta ZORDER BY (InvoiceNo)
```

> metrics (check above cell result)

- numFilesAdded: 1  
After the operation, 1 new file was created to replace existing files. (We will verify this below)

- numFilesRemoved: 16  
16 old files were removed as part of the optimization process.

```python
%sql
SELECT MIN(InvoiceNo), MAX(InvoiceNo), _metadata.file_name 
FROM salesdelta 
GROUP BY _metadata.file_name
ORDER BY MIN(InvoiceNo)

-- After Z-Ordering Optimization
```

```python
%sql
SELECT DISTINCT _metadata.file_name FROM salesdelta;
```

Since we see only 1 partition now, let's verify if the data count is matching before and after Z-Ordering (To ensure there was no DATA Pruning)

```python
%sql
SELECT _metadata.file_name, 
      COUNT(*) AS record_count
FROM salesdelta
GROUP BY _metadata.file_name
ORDER BY record_count DESC;
```

### MultiDimensional Z-Ordering (Adding more than one column for Partition)

```python
%sql
OPTIMIZE salesdelta ZORDER BY (Country, invoiceno)
```

Here, there wont be any change because:
- numFilesAdded: 0
- numFilesRemoved: 0

```python
%sql
SELECT country, MIN(InvoiceNo), MAX(InvoiceNo), _metadata.file_name 
FROM salesdelta 
GROUP BY country, _metadata.file_name
ORDER BY country, MIN(InvoiceNo)
```

```python
%sql
-- Running thi without Country to check what happens in SPARK UI DAG 
SELECT * FROM salesdelta WHERE InvoiceNo='540646' 
```

- In SPARK UI DAG, the above execution will show 'number of files read' as 2, one for country and one for invoice no (if some files were added or removed during OPTIMIZE session)
- We should specify both columns we used in OPTIMZE ZORDER, then it will show 'number of files read' as 1 now which is ideal!!!

```python
%sql
-- Running thi without Country to check what happens in SPARK UI DAG 
SELECT * FROM salesdelta WHERE InvoiceNo='536389' and Country='Australia'
```

### Combination of Z-Ordering on one column and Partitioning on another (Much better)

```python
df = spark.read.csv("/databricks-datasets/definitive-guide/data/retail-data/all/online-retail-dataset.csv", inferSchema=True, header=True)
df.repartition(16).write.format("delta").mode("overwrite").partitionBy("Country").option("path","/data/output/sales_delta_partitioned/").saveAsTable("salesdeltapartitioned")
```

```python
display(dbutils.fs.ls("/data/output/sales_delta_partitioned/"))
```

```python
display(dbutils.fs.ls("/data/output/sales_delta_partitioned/Country=Australia/")) # Number of file in each partition is 16
```

```python
%sql
-- To check performance first without OPTIMIZATION in Spark UI
SELECT * FROM salesdeltapartitioned WHERE InvoiceNo='536389' and Country='Australia'
```

```python
%sql
OPTIMIZE salesdeltapartitioned  WHERE country='Australia' ZORDER BY (invoiceno)
```

```python
%sql
-- To check performance first AFTER OPTIMIZATION, it will take less time to execute now
SELECT * FROM salesdeltapartitioned WHERE InvoiceNo='536389' and Country='Australia'
```
