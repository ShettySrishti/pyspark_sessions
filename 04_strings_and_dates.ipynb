```python
# Create emp DataFrame
emp = spark.createDataFrame(data,schema)
emp.show()

#Schema for emp  
emp.printSchema() 
```

```python
# Case when 
# Select employee_id, name, age, salary, gender
# case when gender = 'Male' then 'M' when gender = 'Female' then 'F' else null end as new_gender, hire_date from emp

from pyspark.sql.functions import when, col, expr
emp_gender = emp.withColumn("new_gender", when(col("gender") == 'Male', 'M')
                            .when(col("gender") == 'Female', 'F')
                            .otherwise(None)

                            #OR

                            )
emp_gender_1 = emp.withColumn("new_gender", expr("case when gender = 'Male' then 'M' when gender = 'Female' then 'F' else null end"))

```

```python
# Replace in strings
# Select employee_id, name, replace(name, 'J', 'Z') as new_name, age, salary, gender, new_gender, hire_date from emp_gender_fixed
from pyspark.sql.functions import regexp_replace

emp_name = emp_gender.withColumn("new_name", regexp_replace(col("name"), 'J', 'Z'))
emp_name.show()
```

```python
# Convert Date
# select *,  to_date(hire_date, 'YYYY-MM-DD') as hire_date from emp_name_fixed

from pyspark.sql.functions import to_date
emp_date_fix = emp_name.withColumn("hire_date", to_date(col("hire_date"), 'yyyy-MM-dd'))
```

```python
#Verifying if the hire_date is converted to 'date' from 'string' datatype 
emp_date_fix.printSchema() 
```

```python
# Add Date Columns
# Add current_date, current_timestamp, extract year from hire_date
from pyspark.sql.functions import current_date, current_timestamp

emp_dated = emp_date_fix.withColumn("date_now", current_date()).withColumn("timestamp_now", current_timestamp())
```

```python
# DROP Null gender records
emp_drop = emp_dated.na.drop()
```

```python
# Fix Null values
# select *, nvl('new_gender', 'O') as new_gender from emp_dated

from pyspark.sql.functions import coalesce, lit
emp_null_df = emp_dated.withColumn("new_gender", coalesce(col("new_gender"), lit("O"))) #wherever the value is null we are denoting it as 'O' (others)
```

```python
emp_null_df.show()
```

```python
# Drop old columns and Fix new column names
# Dropping name, gender and renaming 'new_name' and 'new_gender' to 'name' and 'gender'respectively
emp_final = emp_null_df.drop("name", "gender").withColumnRenamed("new_name", "name").withColumnRenamed("new_gender", "gender")
emp_final.show()
```

```python
# Write data as CSV
emp_final.write.format("csv").save("data/output/4/emp.csv")
```
