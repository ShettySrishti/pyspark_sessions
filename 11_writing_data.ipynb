```python
# Create emp dataframe
emp = spark.createDataFrame(data,schema)
```

```python
# Get number of partitions and show data
emp.rdd.getNumPartitions() 
emp.show()
```

```python
# Writing the data in parquet format
emp.write.parquet("data/output/11/2/emp.parquet")
```

```python
# View data partition information
from pyspark.sql.functions import spark_partition_id
emp.withColumn('partition_id', spark_partition_id()).show()
```

```python
emp.write.option("header",True).csv("data/output/11/3/emp.csv")
```

Write modes - overwrite, append, ignore and error
- append --> add data to the old data
- ignore --> It will do nothing if the required data isnt available
- error --> It will throw and error if the required data is already present

```python
emp.write.partitionBy("department_id").option("header", True).mode("overwrite").csv("data/output/11/4/emp.csv")
```
