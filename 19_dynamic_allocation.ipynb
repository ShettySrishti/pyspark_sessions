```python
# Spark Session
from pyspark.sql import SparkSession
spark = (
        SparkSession
        .builder
        .appName("Dynamic Allocation")
        .config("spark.executor.cores",2)
        .config("spark.executor.memory", "512M")
        .config("spark.dynamicAllocation.enabled", True)
        .config("spark.dynamicAllocation.minExecutors",0)
        .config("spark.dynamicAllocation.maxExecutors",5)
        .config("spark.dynamicAllocation.initialExecutors",1)  # Spark will allocate one executor for this app
        .config("spark.dynamicAllocation.shuffleTrackingEnabled", True)
        .config("spark.dynamicAllocation.executorIdleTimeout","60s") # To scale down the executors --> The executor will be killed by the driver if its not used after 60 seconds
        .config("spark.dynamicAllocation.cacheExecutorIdleTimeout","60s") # To scale down the executors
        .getOrCreate()
)
spark
```

```python
# Read Sales CSV file with 1 million records
sschema ="transacted_at string, trx_id string, retailer_id string, description string, amount double, city_id string"
sales = spark.read.schema(sschema).option("header",True).csv("/data/input/datasets/sales.csv")
```

```python
# Read Cities CSV file with 5 million records
cschema ="city_id string, city string, state string, state_abv string, country string"
cities = spark.read.schema(cschema).option("header",True).csv("/data/input/datasets/cities.csv")
```

```python
# JOIN data
dfjoinedsales = sales.join(cities, 'city_id', 'left_outer')
```

```python
dfjoinedsales.write.format("noop").mode("overwrite").save()  # For performance benchmarking -- Check Spark UI for more info 
```
